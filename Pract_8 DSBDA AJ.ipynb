```python
# Import pandas, numpy  and seaborn package 
import pandas as ps
import numpy as np
import seaborn as sns
```

```python
#Reading csv to dataframe
dataset=sns.load_dataset("titanic")
```

```python
#dimension of dataset
dataset.shape
```

```python
#datatypes of dataset
dataset.dtypes
```

```python
#finding missing values and their count
dataset.isnull().sum()
```

```python
print(dataset)
```

```python
#remove deck column using drop function
dataset.drop('deck',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
#copy of dataset
x=dataset
x1=x
x1.dropna()
```

```python
#Scatter plots are used to observe relationships between variables here variables are age and fare 
sns.scatterplot(x='age',y='fare',data=x1)
```

```python
#hue -  goups variable that will produce points with different colors. e.g blue for first, orange for second and green for third
sns.scatterplot(x='age',y='fare',hue='class',data=x1)
```

```python
#In a histogram, each bar groups numbers into ranges. 
sns.histplot(x1,x='fare')
```

```python
# Each bin is plotted as a bar whose height corresponds to how many data points are in that bin
sns.histplot(x1,x='fare',hue='pclass',bins=10)
```

```python
#kde - used to smooth a histogram.
sns.histplot(x1,x='fare',hue='pclass',bins=10,kde=True)
```

```python
#distplot() is used to visualize the parametric distribution of a dataset.
sns.displot(x1['fare'],aspect=3)
```

```python
#distplot() is used to visualize the parametric distribution of a dataset.
sns.displot(x1['age'],aspect=2)
```

```python
sns.displot(x1,x='age',hue='sex',aspect=2)
```

```python
#col Variables that define subsets to plot on different facets.
sns.displot(x1,x='age',hue='sex',col='pclass')
```

```python
#kde - used to smooth a displot.
sns.displot(x1,x='age',hue='sex',col='pclass',kde=True)
```

```python
sns.histplot(x1['class'])
```

```python
#The barplot() is used to display the mean value for each value in a categorical column, against a numeric column.
sns.barplot(x='sex',y='age',data=x1)
```

```python
sns.barplot(x='sex',y='age',hue='embarked',data=x1,ci=None)
```

```python
#countplot displays the count of the categories in a specific column.
sns.countplot(x='pclass',data=x1)
```

```python
sns.countplot(x='pclass',hue='sex',data=x1)
```

```python
#count of people survived from classes
sns.countplot(x='survived',hue='pclass',data=x1)
```

```python
#Bar plot- to visualize passenger based on class and age along with the place from which passengers embarked(Boarded in ship)
sns.barplot(x = 'age',
            y = 'fare',
            hue = 'pclass',
            data = x1,
            palette = "Blues")
```

```python
#lineplot is a graph that displays data as points or check marks above a number line, 
#showing the frequency of each value.
sns.lineplot(x='age',y='fare',data=x1)
```

```python
#pair plot-To plot multiple pairwise bivariate distributions in a dataset

import matplotlib.pyplot as plt
sns.pairplot(x1,hue='pclass')
```
