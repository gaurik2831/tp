### Dataset : dirtydata

```python
#Reading csv to dataframe
df=pd.read_csv("dirtydata1.csv")
df
```

```python
#finding Missing Values and their count 
nv=df.isnull().sum()
print(nv)
```

```python
# dataframe shape
df.shape
```

```python
#description of the data
df.describe()
```

```python
#data types in the DataFrame
df.dtypes
```

```python
#abs(): to remove the -ve values
df["Calories"]=df["Calories"].abs()
```

```python
df
```

```python
#mean of calories
x=df['Calories'].mean()
x
```

```python
#filling null values of calories column with mean value
df['Calories'].fillna(x,inplace=True)
```

```python
#convert data type of calories from float to integer
df["Calories"]=df["Calories"].astype(int)
df
```

```python
df
```

```python
#removes all the rows of Date column that contains NULL values
df.dropna(subset='Date',inplace=True)
```

```python
df
```

```python
#to convert string Date time into Python Date time object
df['Date']=pd.to_datetime(df['Date'])
df
```

```python
#here we select the location and set the values
df.loc[7,'Duration']=45
```

```python
df
```

```python
#check for duplicate value
df.duplicated()
```

```python
#total numbers of duplicate rows
df.duplicated().sum()
```

```python
#remove duplicate data
df.drop_duplicates(inplace=True)
df
```

```python
#to save file after preprocessing
df.to_csv("dirty_preposseddata.csv")
```

```python
df_nba.isnull().sum()  #to check the null values columnwise and find their count
```

```python
y=df_nba['Salary'].mean()  #find mean of Salary column
print(y)
```

```python
df_nba['Salary'].fillna(780000,inplace=True)     #filling NaN values in salary with constant value
#The fillna() method replaces the NULL values with a specified value
#inplace=True keyword in a pandas method changes the default behaviour
df_nba
```

```python
df_nba['Salary']=df_nba['Salary'].abs()   #to convert negative salary value to positive
```

```python
df_nba
```

```python
df_nba['Position'].unique()   #finding unique values of Position column
```

```python
df_nba.duplicated().sum()
```

```python
df_nba['Position'].value_counts()
```

```python
df_nba['Pos']=df_nba['Position'].replace(['SG','PF','PG','SF','C'],[1,2,3,4,5])     #categorical to quantative 
```

```python
df_nba
```

```python
#categarial to quantitative using label encoding
from sklearn import preprocessing
l_en=preprocessing.LabelEncoder()
```

```python
df_nba['Position']=l_en.fit_transform(df_nba['Position'])  #Fit label encoder and return encoded labels
print(df_nba)
```

```python
df_nba['Age'].unique()  #finding unique values of Age column
```

```python
#quatitative to categarical in python using pandas
category=pd.cut(df_nba.Age,bins=[19,25,30,35,45],labels=['A','B','C','D'])
print(category)
```

```python
#to insert new column in existing dataset
df_nba.insert(5,"Age_group",category)  #here 5 is the column position
df_nba
```
